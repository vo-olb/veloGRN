```python
import numpy as np
import matplotlib.pyplot as plt

# 
def F(x, y):
    return (x**4/(1+x**4) + 1/(1+y**4) - x, 
            y**4/(1+y**4) + 1/(1+x**4) -y)

# 
x = np.linspace(0, 3, 40)
y = np.linspace(0, 3, 40)
X, Y = np.meshgrid(x, y)

# 
U, V = F(X, Y)

# 
plt.figure(figsize=(5,5))
plt.quiver(X, Y, U, V, scale=30)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```
